### Is there any relationship between sleep quality and students attention levels?

```python
plt.figure(figsize=(7,5))
sns.lineplot(x='Kindly Rate your Sleep Quality üò¥', y='How would you rate your stress levels?', data=df)
plt.title("Relationship between Sleep Quality and Stress Levels")
plt.xlabel("Sleep Quality (1=Poor, 5=Excellent)")
plt.ylabel("Stress Level (1=Low, 5=High)")
plt.show()

corr_sleep_stress = df['Kindly Rate your Sleep Quality üò¥'].corr(df['How would you rate your stress levels?'])
print(f"Correlation between Sleep Quality and Stress: {corr_sleep_stress:.2f}")
```

### Are frequent headaches related to high stress levels?

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='How many times a week do you suffer headaches ü§ï?', y='How would you rate your stress levels?', data=df, palette='coolwarm')
plt.title("Headache Frequency vs Stress Levels")
plt.xlabel("Headache Frequency per Week")
plt.ylabel("Stress Level")
plt.show()

corr_headache_stress = df['How many times a week do you suffer headaches ü§ï?'].corr(df['How would you rate your stress levels?'])
print(f"Correlation between Headache and Stress: {corr_headache_stress:.2f}")
```

### Does study load affect stress levels?

```python
plt.figure(figsize=(7,5))
sns.lineplot(x='how would you rate your study load?', y='How would you rate your stress levels?', data=df, marker='o')
plt.title("Study Load vs Stress Levels")
plt.xlabel("Study Load (1=Light, 5=Heavy)")
plt.ylabel("Stress Level")
plt.show()

corr_studyload_stress = df['how would you rate your study load?'].corr(df['How would you rate your stress levels?'])
print(f"Correlation between Study Load and Stress: {corr_studyload_stress:.2f}")
```

### Do extracurricular activities reduce stress?

```python
plt.figure(figsize=(7,5))
sns.barplot(x='How many times a week you practice extracurricular activities üéæ?', 
            y='How would you rate your stress levels?', data=df, ci=None, palette='viridis')
plt.title("Extracurricular Activities vs Stress Level")
plt.xlabel("Activities per Week")
plt.ylabel("Average Stress Level")
plt.show()
```

### Is high academic performance linked to lower stress?

```python
plt.figure(figsize=(7,5))
sns.regplot(x='How would you rate you academic performance üë©‚Äçüéì?', 
            y='How would you rate your stress levels?', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title("Academic Performance vs Stress Levels")
plt.xlabel("Academic Performance (1=Low, 5=High)")
plt.ylabel("Stress Level")
plt.show()

corr_academic_stress = df['How would you rate you academic performance üë©‚Äçüéì?'].corr(df['How would you rate your stress levels?'])
print(f"Correlation between Academic Performance and Stress: {corr_academic_stress:.2f}")
```
